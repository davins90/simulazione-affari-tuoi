```python
import random
import statistics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.io as pio
import plotly.express as px

pio.renderers.default = 'iframe'
```

```python
def soglia_accettazione(turno):
    # Soglia di accettazione statistica come prima
    if turno < 10:
        media = 0.85
    elif 10 <= turno < 17:
        media = 0.8 - 0.2 * ((turno - 10) / 7.0)
    else:
        media = 0.5

    std_dev = 0.05
    soglia = random.gauss(media, std_dev)
    soglia = max(0.4, min(0.9, soglia))
    return soglia

def offerta_del_dottore_questa_mano(turno):
    # Definisce in quali turni il Dottore fa un'offerta
    # Inizialmente ogni 4 turni (4,8), dal 10 al 15 ogni 3 turni, dopo 15 ogni 2 turni.
    
    if turno < 10:
        # Offer ogni 4 turni
        return (turno % 4 == 0)
    elif 10 <= turno <= 15:
        # Offer ogni 3 turni
        # Un modo semplice: turni 12 e 15
        return (turno == 12 or turno == 15)
    else:
        # Dopo il 15, offerta ogni 2 turni
        return (turno % 2 == 0)

def simula_gioco(num_simulazioni=10_000, seed=42):
    random.seed(seed)
    
    premi_iniziali = [
        0, 1, 5, 10, 20, 50, 75, 100, 200, 500, 5000, 10000, 15000, 20000, 
        30000, 50000, 75000, 100000, 200000, 300000
    ]

    risultati = []
    
    for _ in range(num_simulazioni):
        premi = premi_iniziali[:]
        random.shuffle(premi)
        
        pacco_concorrente = premi.pop()
        pacchi_rimanenti = premi[:]
        
        turno = 0
        premio_finale = None
        gioco_finito = False
        
        while not gioco_finito:
            turno += 1
            
            # Controlla se c'Ã¨ un'offerta del Dottore
            if offerta_del_dottore_questa_mano(turno):
                tutti_i_premi = pacchi_rimanenti + [pacco_concorrente]
                media_premi = statistics.mean(tutti_i_premi)

                fattore = random.uniform(0.4, 0.6)
                offerta = media_premi * fattore

                soglia = soglia_accettazione(turno)

                if offerta > soglia * media_premi:
                    premio_finale = offerta
                    gioco_finito = True
                else:
                    # Rifiutata offerta: elimina un pacco
                    if len(pacchi_rimanenti) > 1:
                        indice_da_eliminare = random.randint(0, len(pacchi_rimanenti)-1)
                        del pacchi_rimanenti[indice_da_eliminare]
                    else:
                        # Se rimane un solo pacco, fine gioco
                        ultimo_pacco = pacchi_rimanenti[0]
                        premio_finale = max(pacco_concorrente, ultimo_pacco)
                        gioco_finito = True
            else:
                # Nessuna offerta: si elimina direttamente un pacco
                if len(pacchi_rimanenti) > 1:
                    indice_da_eliminare = random.randint(0, len(pacchi_rimanenti)-1)
                    del pacchi_rimanenti[indice_da_eliminare]
                else:
                    ultimo_pacco = pacchi_rimanenti[0]
                    premio_finale = max(pacco_concorrente, ultimo_pacco)
                    gioco_finito = True

        risultati.append({
            "premio_finale": premio_finale,
            "turno_finale": turno,
            "valore_pacco_concorrente": pacco_concorrente
        })
    
    df = pd.DataFrame(risultati)
    return df
```

```python
# Eseguiamo la simulazione
df_risultati = simula_gioco()
```

```python
# Creiamo un grafico per mostrare l'andamento della distribuzione del premio finale in funzione del turno finale.
# Ad esempio, possiamo mostrare i quantili 10%, 25%, 50%, 75%, 90% come linee separate per ciascun turno.

quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]
grouped = df_risultati.groupby("turno_finale")["premio_finale"].quantile(quantiles).unstack()

plt.figure(figsize=(10, 6))

for q in quantiles:
    plt.plot(grouped.index, grouped[q], label=f'{int(q*100)}Â° percentile')

plt.title("Distribuzione del premio finale per turno di conclusione")
plt.xlabel("Turno Finale")
plt.ylabel("Premio Finale (â¬)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calcolo dei quantili del premio in base al turno finale
quantiles = [0.1, 0.25, 0.5, 0.75, 0.9]
grouped = df_risultati.groupby("turno_finale")["premio_finale"].quantile(quantiles).unstack()

turni = grouped.index
q10 = grouped[0.1]
q25 = grouped[0.25]
q50 = grouped[0.5]
q75 = grouped[0.75]
q90 = grouped[0.9]

# Creazione del grafico con Plotly
fig = go.Figure()

# Aggiungiamo i quantili dall'alto verso il basso creando fasce di colore
# Ordine: q90 (nessun fill), poi q75 (fill a tonexty), q50, q25, q10
# In questo modo creiamo un effetto fan con sfumature.
fig.add_trace(go.Scatter(
    x=turni, y=q90, 
    mode='lines', 
    line=dict(color='rgba(0,100,250,0.2)', width=2), 
    name='90Â° percentile'
))

fig.add_trace(go.Scatter(
    x=turni, y=q75,
    mode='lines',
    line=dict(color='rgba(0,100,250,0.3)', width=2),
    fill='tonexty', # riempie verso il trace precedente (q90)
    name='75Â° percentile'
))

fig.add_trace(go.Scatter(
    x=turni, y=q50,
    mode='lines',
    line=dict(color='rgba(0,100,250,0.4)', width=2),
    fill='tonexty', # riempie verso il trace precedente (q75)
    name='50Â° percentile (median)'
))

fig.add_trace(go.Scatter(
    x=turni, y=q25,
    mode='lines',
    line=dict(color='rgba(0,100,250,0.5)', width=2),
    fill='tonexty', # riempie verso il trace precedente (q50)
    name='25Â° percentile'
))

fig.add_trace(go.Scatter(
    x=turni, y=q10,
    mode='lines',
    line=dict(color='rgba(0,100,250,0.6)', width=2),
    fill='tonexty', # riempie verso il trace precedente (q25)
    name='10Â° percentile'
))

fig.update_layout(
    title="Distribuzione del premio finale per turno di conclusione",
    xaxis_title="Turno Finale",
    yaxis_title="Premio Finale (â¬)",
    hovermode="x unified"
)

fig.show()
```
